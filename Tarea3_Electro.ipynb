```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Vx_1: Potencial en x=0
# Vx_2: Potencial en x=a
# Vy_1: Potencial en y=0
# Vy_2: Potencial en y=b
# N: Número de puntos en los que se va a dividir la base x
# M: Número de puntos en los que se va a dividir la base y
# d: Distancia de paso (distancia entre cada punto de la red asumiendo dx=dy)


def potencial_relajacion(Vx_1, Vx_2, Vy_1, Vy_2, N, M,d):
    
    # Se calcula la posición final en "x" y "y" con la distancia de paso y el número de puntos en cada base 
    a = N*d
    b = M*d
    
    # Se crea la base de la red
    x = np.linspace(0,a,N)
    y = np.linspace(0,b,M)
    
    # Se crea la matriz de la red 
    potencial = np.zeros((N,M))
    
    # Se establecen las condiciones de frontera en los bordes de la red 
    potencial[0,:] = Vx_1
    potencial[-1,:] = Vx_2
    potencial[:,0] = Vy_1
    potencial[:,-1] = Vy_2
    
    # Cantidad de iteraciones que se van a realizar 
    iteraciones = int(1e4)
  
    # Diferencia entre el valor promedio del voltaje de la anterior iteracion y la nueva iteracion 
    diferencia_esperada = 0.

    # Tolerancia del cambio de Voltaje para truncar la relajación, el procedimiento se trunca cuando 
         # los valores del potencial no cambien sensiblemente 
    tolerancia = 1e-6
    
    for iteraciones in range(iteraciones):
        
        for i in range(1, len(x)-1):
            for j in range(1, len(y)-1):
                
                # Se encuentra el voltaje promedio de los puntos vecinos
               potencial_promedio = ( potencial[i+1,j] + potencial[i-1,j] + potencial[i,j+1] + potencial[i,j-1] )/4
                
                # Se calcula la diferencia entre el valor promedio del potencial de la anterior relajación y el nuevo promedio
               diferencia_real = potencial_promedio - potencial[i,j]
                
                # Se actualiza el valor del potencial en el punto en la matriz
               potencial[i,j] = potencial_promedio
                
                # Se calcula y se actualiza la diferencia entre los valores promedios del potencial
               if np.abs(diferencia_real) > diferencia_esperada:
                diferencia = diferencia_real

        # Terminar el procedimiento si la diferencia entre los valores promedios del potencial es menor a la tolerancia
        if np.abs(diferencia) < tolerancia:
            break
    
    #Devuelve la solución del potencial V y las bases "x" y "y" para graficar
    return potencial, x, y, a, b
```

```python
# Valores establecidos para probar el metodo de relajacion 
Vx_1 = 0
Vx_2 = 0
Vy_1 = -5
Vy_2 = 5
N = 100
M = 100
d = 0.1

# Calculo del potencial a partir de la solución computacional
potencial_computacional, x, y, a, b = potencial_relajacion(Vx_1, Vx_2, Vy_1, Vy_2, N, M, d)

# Funcion para calcular el potencial a partir de la solución analítica encontrada
def potencial_analitico(x,y,a,b, V_0):
    
    # Se crean las bases de los valores "x" y "y" que se van a utilizar y ademas se crea laba para guardar los 
         # resultados del potencial en cada punto
    X,Y = np.meshgrid(x,y)
    V = np.zeros_like(X)
    
    # Se hace una sumatoria desde 1 hasta n = 200 tomando solo los valores impares ya que en los valores pares el resultado es cero
    for n in range(1,200,2):

        # Se calcula la constante Bn
        Bn = (1/np.tanh(n*np.pi*b/a))+(1/np.sinh(n*np.pi*b/a))

        # Se calcula el valor del potencial haciendo la suma de los n's en la respectiva solucion encontrada previamente 
        V += (np.sin(n*np.pi*X/a)/n)*(Bn*np.sinh(n*np.pi*Y/a) - np.cosh(n*np.pi*Y/a))
                                                                
        
    return (4*V_0/np.pi)*V

# Potencial a partir de la solucion analítica
V = potencial_analitico(x,y,a,b,5)
```

```python
# Codigo para graficar la solucion numerica (metodo de relajacion) de la ecuacion de Laplace
fig = plt.figure(figsize=(15,10))

X,Y = np.meshgrid(x,y)

ax = fig.add_subplot(121)
ax.contourf(Y,X, potencial_computacional, cmap='inferno')
ax.set_xlabel("x")
ax.set_ylabel("y")

ax1 = fig.add_subplot(122, projection='3d')
ax1.plot_surface(Y,X,potencial_computacional, cmap='inferno')
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("Potencial V")
ax1.set_title("Potencial método de relajación")
ax1.view_init(elev=30, azim=220)

```

```python
# Codigo para graficar la solucion analitica de la ecuacion de Laplace 

fig = plt.figure(figsize=(15,10))

X,Y = np.meshgrid(x,y)

ax = fig.add_subplot(121)
ax.contourf(X, Y, V, cmap='plasma')
ax.set_xlabel("x")
ax.set_ylabel("y")

ax1 = fig.add_subplot(122, projection='3d')
ax1.plot_surface(X,Y ,V, cmap='plasma')
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("Potencial V")
ax1.set_title("Potencial método analítico")
ax1.view_init(elev=30, azim=220)

```

```python
# Codigo para graficar el error entre los dos metodos 

fig = plt.figure(figsize=(15,10))

X,Y = np.meshgrid(x,y)

ax1 = fig.add_subplot(122, projection='3d')
ax1.plot_surface(X,Y ,V-potencial_computacional.T, cmap='cool')
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("Potencial V")
ax1.set_title("Error entre los métodos")
ax1.view_init(elev=30, azim=220)


diferencia = np.abs(V-potencial_computacional.T)
print('La diferencia absoluta media entre los dos metodos es: ' + str(np.round(np.mean(diferencia),2)))
print('la maxima diferencia entre los dos metodos es: '+ str(np.max(diferencia)))
```
